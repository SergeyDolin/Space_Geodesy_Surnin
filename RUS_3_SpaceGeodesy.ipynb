# Тема 3 Измерение времени распространения сигнала

В настоящее время существует 2 способа измерения времени распространения сигнала: запросный и беззапросный 

В **запросном** способе генерируемое время посылается вместе с электромагнитным сигналом на КА. Посылаемый сигнал отражается (лазерным дальномером) или ретранслируется (радиодальномером) на НП.  


**Основные этапы метода:** (дополнение Долин)
1. **Запрос от приемника**: Приемник отправляет запрос на спутник. Время отправки $(T_1)$ фиксируется по часам приемника.

2. **Обработка запроса спутником**: 
   - Спутник принимает сигнал в момент $(T_2)$ (по своим часам). 
   - Формирует ответный сигнал, который может включать метку времени $(T_2)$. 
   - Отправляет ответ в момент $(T_3)$. Задержка обработки на спутнике: $(D = T_3 - T_2)$.

3. **Прием ответа**: 
   - Приемник получает ответ в момент $(T_4)$ (по своим часам). 
   - Измеряется общее время цикла "туда-обратно": $(\Delta T = T_4 - T_1 )$.

4. **Расчет времени распространения**:
   - Если задержка обработки $(D)$ известна или пренебрежимо мала, время в одну сторону: $( \tau = \frac{\Delta T - D}{2} )$.
   - Расстояние: $( R = c \cdot \tau )$, где $( c )$ — скорость света.

## 3.1 Понятие о геометрической дальности и псведодальности

И в том и в другом случае время измерения вычисляется как $\tau = t_{приема} - t_{отправки}$ $(3.1)$

По скорости распространения электромагнитного сигнала (скорость света) $c$ вычисляется дальность $\rho$. 
- в запросном способе $\rho = \frac{\tau}{2}*c$ $(3.2)$
- в беззапросном способе $\rho = \tau*c$ $(3.3)$

Формулы $(3.2)$ и $(3.3)$ отражают суть 2-х способов получения геометрического расстояния $\rho$ по времени распространения электромагнитного сигнала $\tau$ в идеальном случае.

В формулы $(3.2)$ и $(3.3)$ вносится ряд поправок.

Рассмотрим схему измерений псевдодальности беззапросной аппаратурой

![alt text](./Рисунки_Pictures/Fig3.3.jpeg "Схема измерений псевдодальности")

Псевдодальность $\tilde{\rho}$ вычислена по формуле $(3.3)$ $\tilde{\rho} = \tau*c = (t_{приема} - t_{отправки}) * c$. Отметим, что в литературе (зарубежной и отечественной) геометрическая дальность $\rho_{геом}$ определяется иначе, чем это делается в наших лекциях.

В лекции геометрическая дальность $\rho_{геом}$ - это мгновенное расстояние между НП и КА в один физический момент времени $t$, который определяется по формуле:

$\rho_{геом}(t) = |\mathbf{r(t)} - \mathbf{R(t)}|$ $(3.4)$

или в координатной форме

$\rho_{геом}(t) = \sqrt{(x(t) - X(t))^2 + (y(t) - Y(t))^2 + (z(t) - Z(t))^2}$ $(3.4*)$

Заметим, что в формулах $(3.4)$ и $(3.4*)$ координаты векторов КА $\mathbf{r}, (t)$ и НП $\mathbf{R} (t)$ **обязательно!** должны быть даны относительно одной и той же системы координат.

В зарубежной литературе геометрическая 
$\rho_{геом}(t) = \sqrt{(x_{G}(t_{отправки}) - X_{G}(t_{приема}))^2 + (y_{G}(t_{отправки}) - Y_{G}(t_{приема}))^2 + (z_{G}(t_{отправки}) - Z_{G}(t_{приема}))^2}$ $(3.5)$

где индекс $G$ обозначает земную систему координат $O,x_G,y_G,z_G$

Запись $(3.5)$ не совсем корректна, так как $(3.5)$ $\rho_{геом}(t) = |\mathbf{r_G(t_{отправки})} - \mathbf{R_G(t_{приема})}| = (x_{G}\mathbf{i_G}(t_{отправки}) + y_{G}\mathbf{j_G}(t_{отправки}) + z_{G}\mathbf{k_G}(t_{отправки})) - (X_{G}\mathbf{i_G}(t_{приема}) +  Y_{G}\mathbf{j_G}(t_{приема}) + Z_{G}\mathbf{k_G}(t_{приема}))$
поскольку орты $\mathbf{i_G}, \mathbf{j_G}, \mathbf{k_G}$ в моменты $t_{приема}$ и $t_{отправки}$ могут быть в разных системах координат (например, если система $G$ вращается или движется). Нельзя напрямую вычитать компоненты векторов, записанные в разных базисах.


## Тема 3.2 Поправка за аберрациюю

Угловая скорость движения наблюдений вокруг оси вращения Земли $\omega_{\bigoplus}$ относительно небесной системы координат $Ox_{\gamma}y_{\gamma}z_{\gamma} = \omega_{\bigoplus} = \frac{360^{\degree}}{24^h} = \frac{15^{\degree}}{1^h}$ 

Линейная скорость движущихся наблюдений по сути параллельно оси $Z$

$\mathbf{V_{\bigoplus} = \omega_{\bigoplus}\times{\mathbf{R}}}$

где 

- $\omega = \omega_{\bigoplus}k$
- $k = Oz_{\gamma}$ - орт оси

Если $\omega_{\bigoplus} = 15 \degree/h, R = 6 * 10^6 м$

$|\mathbf{V_{\bigoplus}}| = \mathbf{V_{\bigoplus}} = |\mathbf{\omega_{\bigoplus}}| * |\mathbf{R}| * \sin{(\hat{\mathbf{\omega_{\bigoplus}}, \mathbf{R}})} = \omega_{\bigoplus} * \underbrace{\mathbf{R} * \sin{\phi}}_{r параллельны} = \omega_{\bigoplus} * r_{пар.}$

$\mathbf{V_{\bigoplus}} = \frac{15^{\degree}*6*10^{6}м}{\rho^{\degree} \approx{60^{\degree}}} = 15*10^5 м/ч = 416 м/с$

Из рисунка $(3.3)$ следует, что $\rho_{геом.} (отправки) \neq \rho_{геом.} (приема) \neq \tilde{\rho}$

Разность между геометрической дальностью и псевдодальностью есть поправка за аберрацию $\Delta{\rho_{абер.}}$

$\Delta{\rho_{абер.}} (t_{отпр.}) = \rho_{геом.} (отпр.) - \tilde{\rho}$ $(3.6)$

$\Delta{\rho_{абер.}} (t_{приема}) = \rho_{геом.} (приема.) - \tilde{\rho}$ $(3.7)$

$\Delta{\rho_{абер.}} (t_{отпр.}) \neq \Delta{\rho_{абер.}} (t_{приема})$ $(3.8)$

$\Delta{\rho_{абер.}} (t_{отпр.}) = \mathbf{\dot{R}} * \mathbf{\rho^{\degree}} (t_{отпр.} - t_{приема})$ $(3.9)$

где 
- $\mathbf{\dot{R}} = \mathbf{V_{\bigoplus}}$ вектор линейной скорости движения наблюдений;
- $\mathbf{\rho^{\degree}}$ - орт топоцентрического вектора спутника.

Произведение скалярное

$\mathbf{\dot{R}} * \mathbf{\rho^{\degree}} = \mathbf{\dot{\rho_{нп}}} $ $(3.10)$

есть скалярное произведение измеренной геометрической дальности $\rho$ под влиянием перемещения наблюдателя $M$

Таким образом $\Delta{\rho_{абер.}} (t_{отпр.}) = \mathbf{\dot{\rho_{нп}}} \tau$, $(3.11)$

где $\tau = t_{отпр.} - t_{приема}$ - время распространения электромагнитного сигнала

$\Delta{\rho_{абер.}} (t_{приема}) \simeq \mathbf{\dot{r}} * \mathbf{\rho^{\degree}} * \tau$

где скалярное произведение

$\mathbf{\dot{r}} * \mathbf{\rho^{\degree}} = \dot{\rho_{КА}}$ $(3.12)$

$\dot{\rho_{КА}}$ - скорость изменения геометрической дальности $\rho$ из-за движения КА по орбите

Сравнивая формулы $(3.11)$ и $(3.9)$ видим, что поправка за аберрацию на моменты приема и отправки различны по величине из-за различия скоростей КА.

В зарубежной литературе поправка $(3.9)$ называется поправкой Саньяка. Поправка за аберрация - есть проявление эффекта специальной теории относительности.

*(комментарий Долин С.В.)*

Эффект Саньяка выражается в виде релятивистской поправки к времени распространения сигнала из-за вращения Земли.  

**Физическая суть эффекта Саньяка в ГНСС**


Когда спутник передаёт сигнал к приёмнику на поверхности Земли, вращение планеты приводит к тому, что точка приёма смещается за время распространения сигнала.

**Уравнение Саньяка в ГНСС**


Разность времени задержки из-за вращения Земли описывается выражением:  

$\Delta t_{\text{Саньяка}} = \frac{2 \cdot \mathbf{\Omega} \cdot \mathbf{A}}{c^2}$  

где:  
- $\mathbf{\Omega}$ — вектор угловой скорости вращения Земли $7.292115 \times 10^{-5} рад/с$,  
- $\mathbf{A}$ — вектор площади, образованный положением спутника и приёмника (площадь проекции на плоскость, перпендикулярную $\mathbf{\Omega})$,  
- $c$ — скорость света.  

**В скалярной форме (для приближённых расчётов):**

$\Delta t_{\text{Саньяка}} \approx \frac{2 \Omega A \cos \theta}{c^2}$

где $\theta$ — угол между нормалью к плоскости "спутник–приёмник–центр Земли" и осью вращения Земли.
